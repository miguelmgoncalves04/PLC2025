# DBPedia: obras de Chuck Berry
<pre>
select ?nome ?desc where { 
    ?s a dbo:MusicalArtist.
    ?s foaf:name "Chuck Berry"@en .
    ?w dbo:artist ?s.
    ?w foaf:name ?nome.
    ?w dbo:abstract ?desc
} LIMIT 1000

</pre>

```python
import re

def tpc3(teste):
    tokens = [
        ('COMMENT',   r'#.*'),                     
        ('NEWLINE',   r'\n'),
        ('SKIP',      r'[ \t]+'),
        ('PREFIX',    r'[a-zA-Z_]\w*:[a-zA-Z_]\w*'), 
        ('SELECT',    r'select\b'),
        ('WHERE',     r'where\b'),
        ('OPTIONAL',  r'optional\b'),
        ('FILTER',    r'filter\b'),
        ('VAR',       r'\?[a-zA-Z_]\w*'),
        ('URI',       r'<[^>]*>'),
        ('INT',       r'\d+'),
        ('STRING',    r'"[^"]*"(@[a-z]+)?'),         
        ('OP',        r'[=!<>]+'),
        ('PUNCT',     r'[{}.;,]'),
        ('ERROR',     r'.'),                         
    ]
    
    reconhecidos = []
    linha = 1
    
    tokenreg = '|'.join(f'(?P<{name}>{pattern})' for name, pattern in tokens)
    val = re.finditer(tokenreg, teste, re.IGNORECASE)
    
    for v in val:
        dic = v.groupdict()
        tipo = None
        valor = v.group()
        for key in dic:
            if dic[key]:
                tipo = key
                break
        if tipo == 'NEWLINE':
            linha += 1
            reconhecidos.append((tipo, valor, linha, v.span()))
        elif tipo not in ('SKIP', 'COMMENT'):  # Ignorar espaços e comentários
            reconhecidos.append((tipo, valor, linha, v.span()))
            
    return reconhecidos


# Exemplo que nos é dado no enunciado do tpc
query = """
# DBPedia: obras de Chuck Berry
select ?nome ?desc where {
  ?s a dbo:MusicalArtist.
  ?s foaf:name "Chuck Berry"@en .
  ?w dbo:artist ?s.
  ?w foaf:name ?nome.
  ?w dbo:abstract ?desc
} LIMIT 1000
"""

# Executar o analisador
tokens = tpc3(query)

# Imprimir resultado
for t in tokens:
    print(t)
```
